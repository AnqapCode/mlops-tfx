```python
import os

# Nama Pipeline dan direktori utama sesuai Kriteria 1
PIPELINE_NAME = "arifcahyo27-pipeline"
PIPELINE_ROOT = os.path.join(PIPELINE_NAME)
METADATA_PATH = os.path.join("metadata", PIPELINE_NAME, "metadata.sqlite")
SERVING_MODEL_DIR = os.path.join("serving_model", PIPELINE_NAME)

# Direktori data dan modul
DATA_ROOT = "data"
MODULES_ROOT = "modules"

# Mendefinisikan lokasi berkas Python yang telah kita buat
TRANSFORM_MODULE_FILE = os.path.join(MODULES_ROOT, "transform.py")
TRAINER_MODULE_FILE = os.path.join(MODULES_ROOT, "trainer.py")
TUNER_MODULE_FILE = os.path.join(MODULES_ROOT, "tuner.py")
```

```python
from modules.components import init_components

# Mengisi parameter fungsi init_components
componentsList = init_components(
    data_dir=DATA_ROOT,
    transform_module=TRANSFORM_MODULE_FILE,
    trainer_module=TRAINER_MODULE_FILE,
    tuner_module=TUNER_MODULE_FILE,
    training_steps=500,   # Model akan belajar sebanyak 500 langkah
    eval_steps=100,       # Model akan dievaluasi sebanyak 100 langkah
    serving_model_dir=SERVING_MODEL_DIR
)
```

```python
from tfx.orchestration import pipeline
from tfx.orchestration.beam.beam_dag_runner import BeamDagRunner
from tfx.orchestration.metadata import sqlite_metadata_connection_config

def init_local_pipeline(components, pipeline_root):
    """Fungsi untuk membungkus komponen menjadi satu kesatuan Pipeline"""
    return pipeline.Pipeline(
        pipeline_name=PIPELINE_NAME,
        pipeline_root=pipeline_root,
        components=components,
        enable_cache=False, # Mengaktifkan cache agar proses ulang lebih cepat
        metadata_connection_config=sqlite_metadata_connection_config(METADATA_PATH),
        beam_pipeline_args=['--direct_num_workers=1'] # Parameter untuk Apache Beam lokal
    )
```

```python
# Menjalankan Pipeline menggunakan BeamDagRunner
BeamDagRunner().run(
    init_local_pipeline(componentsList, PIPELINE_ROOT)
)
```
