```python
import requests
import json

# URL server lokal TensorFlow Serving kita
# format: http://<host>:<port>/v1/models/<model_name>:predict
URL = "http://localhost:8501/v1/models/spam_model:predict"

# Menyiapkan data SMS percobaan
# Karena model kita membaca input dari fitur 'text_xf', kita format seperti ini:
data = {
    "instances": [
        {"text_xf": "selamat! anda memenangkan hadiah uang tunai 100 juta rupiah. klik link berikut untuk klaim."}, # Pesan Spam
        {"text_xf": "halo ibu, hari ini aku pulangnya agak telat ya karena ada tugas kelompok."} # Pesan Normal (Ham)
    ]
}

# Mengirim data ke server menggunakan metode POST
print("Mengirim request ke server...")
response = requests.post(URL, data=json.dumps(data))

# Menampilkan hasil tebakan model
if response.status_code == 200:
    print("\nBerhasil! Berikut adalah probabilitas prediksi dari model:")
    print(response.json())
else:
    print(f"Gagal memanggil API. Kode Error: {response.status_code}")
    print(response.text)
```
